```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
class AuxCls():
    
    def __init__(self):
        self.alpha = 100
        self.tau = 5*60
        self.deltaT = 60*random.randint(1, 5)
        self.padding = 0
    
    def fun(self, t):
        
        if t >= self.deltaT:
            self.padding = self.deltaT
            self.deltaT += 60*random.randint(5, 10)
       
        return self.alpha*np.exp(-(t - self.padding)/self.tau)
    
```

```python
def drawThis(show=True, export=False):

    W = AuxCls()

    fig = plt.figure(figsize=(8, 6))

    plt.plot(
        time := np.linspace(0, 120, 10000),
        list(map(lambda t: 100*W.fun(t)/W.alpha, 60*time)),
        "-k",
        linewidth=2
    )

    plt.xlim([0, 60])
    plt.ylim([0, 100])

    plt.ylabel("stake power %")
    plt.xlabel("time (min)")
    
    plt.title(
        "Noo mi Pana donde te sentaste!"
    )
    
    if show:
        plt.show()
    
    if export:
        fig.savefig(
            "POC.pdf",
            dpi=400,
            bbox_inches='tight'
        )

```

```python
drawThis(export=True)
```
